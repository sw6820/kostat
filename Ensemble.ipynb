```python
def ensemble(lst):
    lst = list(map(str, lst))
    count = 0
    mode = 0;
    for x in lst: 
        if lst.count(x) > count:
            count = lst.count(x)
            mode = x

    return mode
```

```python
def per(x, percent):
    x_list = []
    for i in range(percent):
      x_list.append(x)
    return x_list
```
